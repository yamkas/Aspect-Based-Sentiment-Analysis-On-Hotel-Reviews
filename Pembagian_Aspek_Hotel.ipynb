```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style("whitegrid")
```

```python
filename = "4000ReviewHotelDprimaTangerang.xlsx"
df = pd.read_excel(filename)
df.head(100)
```

```python
import re

# Daftar kata kunci aspek
aspek_kamar = ["kamar", "ranjang", "kasur", "tempat tidur", "toilet", "ac", "bathroom", "shower", "lemari", "jendela", "lampu", "tv"
]
aspek_pelayanan = ["pelayanan", "staf", "service", "resepsionis", "ramah", "helpful", "sopan", "cepat tanggap", "slow response", "senyum","staff", "stafff"]
aspek_kebersihan = ["bersih", "kebersihan", "higienis", "rapih", "debu", "bau", "kotor", "sampah", "toilet bersih", "handuk kotor"]
aspek_fasilitas = ["wifi", "tv", "kolam renang", "gym", "lift", "sarapan", "breakfast", "parkir", "lounge", "air panas", "shuttle", "bar"
]

# Fungsi untuk mengecek aspek dalam ulasan
def tentukan_aspek(ulasan, keywords):
    return any(re.search(rf"\b{kw}\b", ulasan, re.IGNORECASE) for kw in keywords)

# Menambahkan kolom aspek berdasarkan kata kunci
df["aspek_kamar"] = df["ulasan"].apply(lambda x: tentukan_aspek(x, aspek_kamar))
df["aspek_pelayanan"] = df["ulasan"].apply(lambda x: tentukan_aspek(x, aspek_pelayanan))
df["aspek_kebersihan"] = df["ulasan"].apply(lambda x: tentukan_aspek(x, aspek_kebersihan))
df["aspek_fasilitas"] = df["ulasan"].apply(lambda x: tentukan_aspek(x, aspek_fasilitas))


df.head()
```

```python
# Mengunduh hasil aspek dalam format Excel
df.to_excel("hasil_aspek_baru.xlsx", index=False)
print("File 'hasil_aspek.xlsx' telah diunduh.")
```
